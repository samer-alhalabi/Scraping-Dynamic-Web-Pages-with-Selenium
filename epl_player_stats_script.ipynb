```python
# importing required packages
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import pandas as pd


def player_stats(player_name):

    player_name = player_name
    
    # adding options on chrome
    chromeOptions = Options()
    chromeOptions.add_argument("--kiosk")

    # navigating to the premier league site
    driver = webdriver.Chrome(options=chromeOptions)
    driver.get("https://www.premierleague.com/players")

    # searching for player
    # web driver wait
    time.sleep(1)
    element = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, "search-input")))

    search_ele = driver.find_element_by_id("search-input")
    search_ele.send_keys(player_name)
    search_ele.send_keys(Keys.RETURN)

    # click on Magnifier
    time.sleep(1)
    driver.implicitly_wait(10)
    click_on_Magnifier = driver.find_element_by_xpath("//div[@class='searchIconContainer searchCommit']").click()

    # clicking on player name
    time.sleep(1)
    driver.implicitly_wait(10)

    # click on player's image (link)
    # click_salah = driver.find_element_by_xpath("//img[@data-player='p118748']").click()

    # click on player's name link
    click_on_link = driver.find_element_by_xpath("//a[@class='playerName']").click()

    # stats button element
    time.sleep(1)
    players_stats = driver.find_element_by_xpath("//a[@href='stats']").click()

    # filter by season
    time.sleep(2)
    season_dropdown = driver.find_element_by_xpath("//div[@aria-labelledby='dd-compSeasons']").click()
    time.sleep(2)
    pick_season = driver.find_element_by_xpath("//li[@data-option-name= '2020/21']").click()    
    time.sleep(2)

    # create variable for the final page targeted
    page_source_stats = driver.page_source

    # loading page source info
    time.sleep(2)
    soup = BeautifulSoup(page_source_stats, "lxml")

    # locating all the stats
    stat_finder = soup.find_all("span", class_="stat")

    # Building The dataset
    records = []
    for stat in stat_finder:
        stat_descr = stat.text[:-1]
        column_name = player_name + " " + "stats" + " " + "for season 2020/21"
    #     stat = stat.find("span").string
        records.append((stat_descr))
        
    time.sleep(5)
    driver.quit()    
    
    df = pd.DataFrame(records, columns=[column_name])
    return df
```

```python
player_stats(input("What is the player name? Note: stats will be pulled for season 2020/21. "))
```
